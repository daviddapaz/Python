```python
from itertools import product 

def shpm_scenarios(**kwargs):
    
    target_list = list(kwargs.values())
    print('selection :', target_list)
    
    # converting the list to boolean
    target_list_binary = [True if (x == 1 or x == 0 ) else False for x in target_list]
    n = len(target_list_binary)
    print('selection converted to boolean :', target_list_binary)
    
    # all the possible scenarios
    options = list(product([0, 1], repeat=n))
    print('All possible options for n =', n, '-->',len(options))

    options_list = [list(x) for x in options]
    
    for option in options_list:
        print(option)
        
    results = []
    
    # selection for target cases only
    for i in options_list:
        temp1 = []

        for x,y in zip(i, target_list):

            temp1.append(x==y)

        if temp1 == target_list_binary:
            results.append(i)
    
    # convert to strings - binary
    results_str = []
    
    for i in range(0,len(results)):
        temp = ''.join([str(x) for x in results[i]])
        results_str.append(temp)
    
    print('')
    print(len(results_str), ' options selected')
    for x in results_str:
        print(x)
    return results_str


shpm_results = shpm_scenarios(key1='', key2=0, key3='', key4='', key5=1, key6=0, key7=0)
```
